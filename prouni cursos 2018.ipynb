```python
# quantidade de cursos por UF
df['uf_busca'].value_counts()
```

```python
# qual o curso com a mensalidade mais alta
df[df['mensalidade'] == df ['mensalidade'].max()]
```

```python
# qual o curso com a mensalidade mais barata
df[df['mensalidade'] == df ['mensalidade'].min()]
```

```python
# qual a mensalidade mais cara por UF
df.groupby('uf_busca')['mensalidade'].max()
```

```python
# qual a mensalidade mais barata por UF
df.groupby('uf_busca')['mensalidade'].min()
```

```python
#mostrar todos os dados dos cursos mais caros de casa UF
df.loc[df.groupby('uf_busca')['mensalidade'].idxmax()]

# o loc serve pra localicar uma linha dentro do dataframe
```

```python
# mensalidade entre 500 e 1000
df[(df['mensalidade'] > 500) & (df['mensalidade'] < 1000)]
```

```python
df[(df['mensalidade'] > 9500) & (df['mensalidade'] < 10000)][['universidade_nome','nome','mensalidade','uf_busca']]
```

```python
# Quais as 5 maiores mensalidades?

df.nlargest(5,'mensalidade')
```
