# Analise dos Dados da Copa do Mundo
Com base na pagina sobre na Wikipedia. Vamos obter os dados sobre o publico pagante em cada final.
link -> https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA

```python
# importando bibliotecas

import pandas as pd
import requests

url = 'https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA'

#requisitar a pagina e guardar a resposta

resposta = requests.get(url)

#Exibir

resposta.text

#obter a resposta em formato HTML

table = pd.read_html(resposta.text)

#exibir resultado
table
```

```python
# Queremos apenas a tabela com os pagantes na final
# esse e a 12 tabela com os dados

# Enta o DataFrame sera apenas da tabela 12

df = table[11]

df.head()
```

```python
# Vamos aos tipos de coluna
df.info()
```

```python
df.columns = ['Edição','Ano','Estádio','Local','Público pagante']

df.head()
```

```python
# Vamos construir um grafico com a evolucao do pulico pagante
#importar uma nova biblioteca

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# configurando o matplotlib

%matplotlib inline
```

```python
#Exibindo o grafico de linhas com marcacao em cada ponto

sns.pointplot(x = 'Ano', y = 'Público pagante', data = df)

plt.xticks(rotation=65)
plt.show()
```
