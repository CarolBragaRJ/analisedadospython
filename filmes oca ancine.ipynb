# Analise de Dados de filmes exibidos no Brasil de 2009 a 2018
Com base nos dados da Ancine (oca.ancine.gov.br)


```python
# importando as bibliotecas
import pandas as pd

#Configurando o matplotlib
%matplotlib inline
```

```python
# Criando DataFrame com base no arquivo em Excel
# Precisamos pular as duas primeiras linhas (skiprows)

df = pd.read_excel('2120.xlsx', skiprows =2, skipfooter =16)

#Exibir as 5 primeiras linhas do DF
df.head()
```

```python
#Exibir as ultimas linhas
df.tail()
```

```python
#Verificar todos os tipos das colunas
df.dtypes
```

```python
# Converter o campo data de lançamento em data
df['Data de lançamento'] = pd.to_datetime(df['Data de lançamento'])
```

```python
# Vamos identificar o problema da coluna data de lançamento
df['Data de lançamento'].value_counts()
```

```python
# Exibir algun filmes com data igual a relancamento
df[df['Data de lançamento'] == 'Relançamento'].head()
```

# Decisao sobre dados corretos e faltantes

```python
# Exibi o tamando do DF
df.shape
```

```python
# Criar um novo DataFrame sem os relancamentos

df_novo = df[df['Data de lançamento'] != 'Relançamento'].copy()
```

```python
df_novo.dtypes
```

```python
# Converter a coluna Data de Lancamento para datetime

df_novo['Data de lançamento'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.dtypes
```

```python
# Quais os anos tiveram mais filmes lançados

df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
# Plotar um grafico dos lancamentos por ano
df_novo['Data de lançamento'].dt.year.value_counts().plot.barh()
```

```python
# Qual o filme com maior renda?

df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
# Tres filmes com maior bilheteria

df_novo.nlargest(3,'Renda (R$) no ano de exibição')
```

```python
# Qual o filme com menor renda?

df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```

```python
# Quantos filmes nacionais e estrangeiros?

df_novo['Nacionalidade da obra'].value_counts()
```
